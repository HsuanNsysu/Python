```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('data.csv')
data.head(10)
```

```python
data1=data.set_index(data['date'])
data1.head(5)
```

```python
data2=data1[['open','high','low','close']]
data2.head(5)
```

```python
data3=data2[['close']]
data3.head(5)
```

```python
data3.describe()
```

```python
data2.plot()
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('TWSE')
plt.show()
```
